```python
keywords = {"int", "float", "if", "else", "while", "return"}

source_code = "int a = 10 ; float b = 20 ; if a > b return a ;"

buffer = source_code

words = buffer.split()

for word in words:
    if word in keywords:
        print("KEYWORD:", word)

    elif word.isdigit():
        print("NUMBER:", word)
```

```python
keywords = {"int", "float", "if", "else", "while", "return"}

source_code = "int a = 10 ; float b = 20 ; if a > b return a ;"

buffer_size = 10

keyword_count = 0
number_count = 0
identifier_count = 0

buffer1 = ""
buffer2 = ""

for i in range(0, len(source_code), buffer_size * 2):

    buffer1 = source_code[i:i + buffer_size]
    buffer2 = source_code[i + buffer_size:i + buffer_size * 2]

    combined_buffer = buffer1 + buffer2

    words = combined_buffer.split()

    for word in words:
        if word in keywords:
            keyword_count += 1
        elif word.isdigit():
            number_count += 1
        elif word.isidentifier():
            identifier_count += 1

print("Keywords:", keyword_count)
print("Numbers:", number_count)
print("Identifiers:", identifier_count)
```
